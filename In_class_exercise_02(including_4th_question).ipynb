```python
import urllib.request
from bs4 import BeautifulSoup
link = "https://www.imdb.com/title/tt1454468/reviews?spoiler=hide&sort=helpfulnessScore&dir=desc&ratingFilter=0"
page = urllib.request.urlopen(link)
soup = BeautifulSoup(page)
print(soup.prettify())
```

```python
soup.find_all("a")
```

```python
#Printing all links
all_links = soup.find_all("a")
for link in all_links:
    print(link.get('href'))
```

```python
right_content = soup.find('div', class_='lister-list')
right_content
```

```python
all_data = []

#Printing all titles, names, and times
for row in right_content.find_all("div", class_="lister-item-content"):
    title = row.find_all('a', class_="title")
    name = row.find_all('span', class_="display-name-link")
    time = row.find_all('span', class_="review-date")
    
    print(title)
    print(name)
    print(time)
```

```python
#Printing all review content
for row in right_content.find_all("div", class_="content"):
  review = row.find_all('div', class_="text show-more__control")
  print(review)
```

```python
#Printing all ratings
for row in right_content.find_all("span", class_="rating-other-user-rating"):
  star = row.find_all('span')
  print(star)
```
