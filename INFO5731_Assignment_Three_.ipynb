```python
# Question 1 (1)
import pandas as pd
df = pd.read_csv('/content/Iphone_11_reviewws.csv', sep='|', encoding='latin-1')
df.head()
```

```python
# natural language processing: n-gram ranking
import re
import unicodedata
import nltk
from nltk.corpus import stopwords
# add appropriate words that will be ignored in the analysis
ADDITIONAL_STOPWORDS = ['covfefe']

import matplotlib.pyplot as plt
```

```python
def basic_clean(text):
  """
  A simple function to clean up the data. All the words that
  are not designated as a stop word is then lemmatized after
  encoding and basic regex parsing are performed.
  """
  wnl = nltk.stem.WordNetLemmatizer()
  stopwords = nltk.corpus.stopwords.words('english') + ADDITIONAL_STOPWORDS
  text = (unicodedata.normalize('NFKD', text)
    .encode('ascii', 'ignore')
    .decode('utf-8', 'ignore')
    .lower())
  words = re.sub(r'[^\w\s]', '', text).split()
  return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
import nltk
nltk.download('wordnet')
import nltk
nltk.download('stopwords')
  
words = basic_clean(''.join(str(df['Review_Description'].tolist())))
```

```python
# Calculating probabilities
from textblob import TextBlob
import nltk
nltk.download('punkt')
with open('/content/Iphone_11_reviewws.csv', 'r', encoding='latin-1') as file:
  data_file = file.read()
bi = TextBlob(data_file)
bigrams = bi.ngrams(2)
gram={ }
probability={ }
for bigram in bigrams:
  bigram=tuple(bigram)
  gram[bigram]=gram.get(bigram,0)+1
for a, b in gram.items():
  probability[a]=b/bi.words.count(a[0])
result = pd.DataFrame.from_dict(probability, orient = 'index', columns=['bigrams_probability'])
result['bigrams_frequency']=[b for b in gram.values()]
result   
```

```python
#Extracting all noun phrases and their probabilitiese
import spacy
from sklearn.feature_extraction.text import CountVectorizer
nlp = spacy.load("en")
file = open('/content/Iphone_11_reviewws.csv', 'r', encoding='latin-1' )
data_file2 = nlp(file.read())
noun_phrases = []
for noun in data_file2.noun_chunks:
  noun_phrases.append(noun.text)
print(noun_phrases)
np_result = pd.DataFrame(noun_phrases, columns = ['noun_phrases'])
vector = CountVectorizer(ngram_range=(3,3), analyzer='word')
transform1 = vector.fit_transform(np_result['noun_phrases'].values.astype('U')) 
frequency = sum(transform1).toarray()[0]
npp_result = pd.DataFrame(frequency, index=vector.get_feature_names(), columns=['noun_phrases_Frequency'])
npp_result['noun_phrases_Probabilities'] = npp_result / npp_result.max()
npp_result
```

```python
import pandas as pd
df = pd.read_csv('/content/Iphone_11_reviewws.csv', sep='|', encoding='latin-1')
df.head()
```

```python
# To build the documents-terms weights (tf*idf) matrix bold text.
tf2 = df.dropna()
tf1 = (tf2['Review_Description'].apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index())
tf1.columns = ['Review_Description','tf']
import numpy as np
#calculating tf*idf
for i,j in enumerate(tf1['Review_Description']):
 tf1.loc[i,'idf'] = np.long(df.shape[0]/(len(df[df['Review_Description'].str.contains(j, regex=False)])))
tf1['tf*idf'] = tf1['tf'] * tf1['idf']
tf1 
```
