```python
#Importing spacy and en_core_web_sm
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()

entity_types = ['PERSON','NORP','FAC','ORG','GPE','LOC','PRODUCT','EVENT','WORK_OF_ART','LAW',
                'LANGUAGE','DATE','TIME','PERCENT','MONEY','QUANTITY','ORDINAL','CARDINAL']

entity_name = ['person','Nationality','Building','Institution','country','location','PRODUCT','EVENT','Title','LAW',
                'LANGUAGE','DATE','TIME','PERCENT','MONEY','QUANTITY','ORDINAL','CARDINAL']
#Defining entity_extraction sentence
def entity_extraction(sentence):
    entity_list = []
    doc = nlp(sentence)
    for ent in doc.ents:
        entity_list.append([ent.text, ent.start_char, ent.end_char, ent.label_])
    
    return entity_list
Data = "Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide."

entity_list = entity_extraction(Data)
print(entity_list)
```

```python

#Dependancy parsing feature extraction
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide.")
for token in doc:
    print(token.text, token.dep_, token.head.text, token.head.pos_,
            [child for child in token.children])
```

```python
import spacy
from spacy.symbols import nsubj, VERB

nlp = spacy.load("en_core_web_sm")
doc = nlp("Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide.")

# Finding a verb with a subject 
verbs = set()
for possible_subject in doc:
    if possible_subject.dep == nsubj and possible_subject.head.pos == VERB:
        verbs.add(possible_subject.head)
#Printing Verbs
print(verbs)
```

```python
#Parts of speech tagging feature extraction
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide.")

for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
             token.is_alpha, token.is_stop)
```

```python
#Bag of words feature extraction
from sklearn.feature_extraction.text import CountVectorizer

#Assigning the attribute values 
bow = CountVectorizer(max_features=50, lowercase=True, ngram_range=(1,3),analyzer = "word") 
Data = ["Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide." ]
train_bow = bow.fit_transform(Data)
#Getting feature names from bag of words of the document
bow.get_feature_names()
```

```python
#tf-idf feature extraction

from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

#Assigning the countvectorizer class functionality
vec = CountVectorizer(binary=True) 
Data = ["Jeff Bezos founded Amazon from his garage in Bellevue, Washington, on July 5, 1994. It started as an online marketplace for books but expanded to sell electronics, software, video games, apparel, furniture, food, toys, and jewelry. In 2015, Amazon surpassed Walmart as the most valuable retailer in the United States by market capitalization. In August 2017, Amazon acquired Whole Foods Market for US$13.4 billion, which substantially increased its footprint as a physical retailer. In 2018, its two-day delivery service, Amazon Prime, surpassed 100 million subscribers worldwide."]

#Fitting the data with the function
vec.fit(Data)
print([w for w in sorted(vec.vocabulary_.keys())])

pd.DataFrame(vec.transform(Data).toarray(), columns=sorted(vec.vocabulary_.keys()))
```

```python
#Importing pandas
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
#Calculating tf-idf values
vec = TfidfVectorizer()
#Assigning the data
vec.fit(Data)
pd.DataFrame(vec.transform(Data).toarray(), columns=sorted(vec.vocabulary_.keys()))
```
